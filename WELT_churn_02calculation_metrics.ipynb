## Metriken zum Angucken der Verteilungen

```python
last_contract_KTag.loc[last_contract_KTag['WT SSOID']=='ASAG5295080']
```

```python
# calculating metrics for inspecting user behaviour
def calc_metrics(dat):
    
   # print(dat.name)
    
    # delete user behavior after cancellation in order to get clean data
    if (last_contract_KTag.loc[last_contract_KTag['WT SSOID']==dat.name,'Kuendigungsstatus'] == 'gekuendigt').bool():
        # cancellation day is K Tag vrom contract data
        canc_day = last_contract_KTag.loc[last_contract_KTag['WT SSOID']==dat.name,'K Tag']
        # if cancellation day is not set, set VE Tag
        if pd.isnull(canc_day).bool():
            canc_day = last_contract_KTag.loc[last_contract_KTag['WT SSOID']==dat.name,'VE Tag']
            
        canc_day = canc_day + timedelta(days=7)
        
        # if cancellation date is before webtrekk-data, delete all entries
        if (canc_day < pd.datetime(2017,9,4)).bool():
            for idx, val in enumerate(dat):
                dat.iloc[idx] = np.nan
        # otherwise delete activity after cancellation day    
        else:
            p = pd.PeriodIndex(freq='w', start=canc_day.dt.date.item(), end=pd.datetime(2018,2,28))
            for idx,val in enumerate(p):
                idxStr = str(val.year) + '/' + str(val.weekofyear)
                dat[idxStr] = np.nan
    
    
    ### STANDARD METRICS
    
    # mean
    mean = dat.mean() 
    # median
    median = dat.median()
    # sd
    sd = dat.std()
    # number of active weeks
    active = len(dat) - sum(dat.isna())
    # min
    mini = dat.min()
    # max
    maxi = dat.max()
    
    # write data
    dat['mean'] = mean
    dat['median'] = median
    dat['sd'] = sd
    dat['activ_wks'] = active
    dat['min'] = mini
    dat['max'] = maxi
    dat['diff_max_min'] = maxi - mini

    
    
    ### USER BEHAVIOR BEFORE CANCELLATION
    
    # find user behaviour before cancellation, if there is no cancellation return last months  
    # if there is no cancellation just use user behaviour of last three months
    if (last_contract_KTag.loc[last_contract_KTag['WT SSOID']==dat.name,'Kuendigungsstatus'] == 'ungekuendigt').bool() :
        
        if sum(~pd.isnull(dat)) < 4:
            
            # user behaviour last 4 weeks
            dat['usr_behav4'] = np.nan
            dat['usr_behav3'] = np.nan
            dat['usr_behav2'] = np.nan
            dat['usr_behav1'] = np.nan
            
        else:    
            
            # get active weeks = non-zero values 
            actWks = np.nonzero(~pd.isnull(dat.values))
            # sample randomly values from active weeks
            rdmSmp = np.random.choice(actWks[0],4)
            
            # user behaviour last 4 weeks
            dat['usr_behav4'] = rdmSmp[0]
            dat['usr_behav3'] = rdmSmp[1]
            dat['usr_behav2'] = rdmSmp[2]
            dat['usr_behav1'] = rdmSmp[3]

    else:
        
        canc_day = last_contract_KTag.loc[last_contract_KTag['WT SSOID']==dat.name,'K Tag']
        
        # cancellaton date is too early, there is not enough information on user behaviour
        # --> setting values to Na
        canc_jahr = canc_day.dt.year.item()
        canc_kw =canc_day.dt.week.item()

        if (canc_jahr <= 2017) & (canc_kw < 39):

            # user behaviour last 4 weeks
            dat['usr_behav4'] = np.nan
            dat['usr_behav3'] = np.nan
            dat['usr_behav2'] = np.nan
            dat['usr_behav1'] = np.nan
      
        
        else:
                    
            # if cancellation day is not set
            if pd.isnull(canc_day).bool():
                canc_day = last_contract_KTag.loc[last_contract_KTag['WT SSOID']==dat.name,'VE Tag']
                canc_jahr = canc_day.dt.year.item()
                canc_kw =canc_day.dt.week.item()
                
            # if cancellation is too recent, substract one week from canc_day
            if (canc_jahr == 2018) & (canc_kw > 9):
                weeks = canc_kw - 9
                canc_day =  canc_day -  pd.to_timedelta(weeks*7, unit='d')
            
            # date series of last 4 weeks before cancellation date
            p = pd.PeriodIndex(freq='w', end=canc_day.dt.date.item(), periods=4)
            
            # construct correct column index from date series
            # user behavior last 4 weeks
            dat['usr_behav4'] = dat[ str(p[3].year)+'/'+str(p[3].weekofyear) ]
            dat['usr_behav3'] = dat[ str(p[2].year)+'/'+str(p[2].weekofyear) ]
            dat['usr_behav2'] = dat[ str(p[1].year)+'/'+str(p[1].weekofyear) ]
            dat['usr_behav1'] = dat[ str(p[0].year)+'/'+str(p[0].weekofyear) ]
            
            
    # metrics to measure change of user behavior before cancellation
    dat['prz_usr_behav4'] = dat['usr_behav4']/median
    dat['prz_usr_behav3'] = dat['usr_behav3']/median
    dat['prz_usr_behav2'] = dat['usr_behav2']/median
    dat['prz_usr_behav1'] = dat['usr_behav1']/median
    
    dat['zsc_usr_behav4'] = (dat['usr_behav4']-mean)/sd
    dat['zsc_usr_behav3'] = (dat['usr_behav3']-mean)/sd
    dat['zsc_usr_behav2'] = (dat['usr_behav2']-mean)/sd
    dat['zsc_usr_behav1'] = (dat['usr_behav1']-mean)/sd
    

    return(dat)
```

```python
# call function
def apply_function_on_table(table, name):
    table1 = table.apply(func=calc_metrics, axis=1)
    
    # change columns names
    names = [name + x for x in table1.columns.values]
    table1.columns = names
    
    return table1
```

```python
# make calculations (~14 min calculation time)
print('PI')
pi_per_ssoid_met = apply_function_on_table( table = pi_per_ssoid_cont, name = 'PI_' )
print('V')
v_per_ssoid_met = apply_function_on_table( table = v_per_ssoid_cont, name = 'V_' )

print('PIPlus')
piPlus_per_ssoid_met = apply_function_on_table( table = piPlus_per_ssoid_cont, name = 'PIPlus_')
print('VPlus')
vPlus_per_ssoid_met = apply_function_on_table( table = vPlus_per_ssoid_cont, name = 'VPlus_')

print('PI/V')
pi_per_v_cont_met = apply_function_on_table( table = pi_per_v_cont, name = 'PI/V_')
print('PIPlus/V')
piPlus_per_v_cont_met = apply_function_on_table( table = piPlus_per_v_cont, name = 'PIPlus/V_')
print('VPlus/V')
vPlus_per_v_cont_met = apply_function_on_table( table = vPlus_per_v_cont, name = 'VPlus/V_')
```

```python
# all tables will be joined
all_v = pi_per_ssoid_met.join(v_per_ssoid_met).join(piPlus_per_ssoid_met).join(vPlus_per_ssoid_met).join(pi_per_v_cont_met).join(piPlus_per_v_cont_met).join(vPlus_per_v_cont_met)

contract_dats =  last_contract_KTag[['WT SSOID', 'VE Jahr/KW', 'Kuendigungsstatus', 'Kuendigungsgrund',
       'Promotion ID', 'Vertragsform', 'Trial Kennzeichen',
       'Vertragsphase', 'VB Tag', 'VB Jahr', 'VB KW', 'VB Jahr/KW',
       'VE Tag', 'VE Jahr', 'VE KW',
       'K Tag']]
contract_dats.index = contract_dats['WT SSOID']

all_v_cont = contract_dats.join(all_v)
```
